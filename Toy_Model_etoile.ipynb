<h1> Description of the toy model </h1>
<hr>

<center>

![image](https://raw.githubusercontent.com/VSCHY/RiverNetwork/master/star_scheme.png)

<em>Figure 1: Description of the HTUs toy model in Star network </em>
</center>

Qin is shared between the "star component"

River dynamic Parameter : 
*   tcst : time constant of the reservoir
*   topoindex

In this model we consider a basic time unit = 1. Then the subdivision of the time unit are defined by subdt. For exemple, subdt = 2 corresponds to two time step inside the basic unit, i.e. a time step of 0.5.

Time parameters :
*   niter : number of iteration
*   subdt : number of subdivision of 1 time unit

Formula to define the discharge out of i : 

$Qi = \frac{V_i}{(tcst*topoindex)}* dt$

The outflow discharge is Qout = Q(j=2)

```
import numpy as np
import matplotlib.pyplot as plt

class experiment:
  def __init__(self, D, V= None, Vstar=None, Qin = 0):
    """
    D : Initialization of the parameters values from the dictionnary
    V : Array with the initial state of the different reservoir (possible to describe as a list)
    Qin : inflow in the first reservoi
    """
    self.tcst = D["tcst"]
    self.topoindex = D["topoindex"]
    self.nhtus = D["nhtus"]
    # 
    self.subdt =D["subdt"]
    # subdt describe the time step, it is the number of division of the basic time step   
    self.dt = 1/self.subdt
    # 
    self.niter = D["niter"] 
    self.f_cst = 1/(self.tcst * self.topoindex)
    self.init_var()
    if V is not None:
        self.V[:] = V
    if Vstar is not None:
        self.Vstar[:] = Vstar
    self.Qin = Qin

  def init_var(self):
    """
    Initialization of volume and discharge for each reservoir
    """
    self.Vstar = np.zeros((self.nhtus))
    self.Qstar = np.zeros((self.nhtus))  
    self.V = np.zeros((2))
    self.Q = np.zeros((2))  
  
  def launch(self):
     """
     Run the experiment, the output are the mean discharge over each basic unit of time
     and for each iteration
     """
     hist_out = np.zeros(self.subdt*self.niter)
     for it in range(self.niter):
        for t in range(self.subdt):
          self.Qstar = self.Vstar*self.f_cst*self.dt
          self.Vstar = self.Vstar-self.Qstar
          self.Q = self.V*self.f_cst*self.dt
          self.V = self.V-self.Q
          hist_out[it * self.subdt + t] = self.Q[-1]
          #
          self.V[0] += np.sum(self.Qstar)
          self.V[1] += self.Q[0]
          self.Vstar[:] += self.Qin
     # Mean de subdt sur les subdt
     hist_out = np.mean(np.reshape(hist_out, (-1,self.subdt)), axis = 1)
     return hist_out

####

def get_param(tcst = 0.1, topoindex = 300, nhtus = 1, subdt = 1, niter = 1000):
    """
    Function to define the Dictionnary parameters.
    This is simplified by using default values.
    """
    D = {"tcst": tcst, "topoindex":topoindex, "nhtus" :nhtus, "subdt":subdt,  "niter":niter}
    return D
```

```
# Experiment 1 
"""
Let's make a try, first experiment to explore the model
explore different nhtus and different subdt with the first intuition of adjusting the volume to the number of HTUs
Let's fill all the "star" HTUs with 1000 m^3 with the linear HTUs empty.
"""
Lout = []; Lname = []

exp0 = experiment(get_param(), Vstar = [1000])
Lout.append(exp0.launch()); Lname.append("nhtus = 1, subdt = 1")

exp0 = experiment(get_param(nhtus = 2), Vstar = [1000, 1000])
Lout.append(exp0.launch()); Lname.append("nhtus = 2, subdt = 1")

exp0 = experiment(get_param(subdt = 3), Vstar = [1000])
Lout.append(exp0.launch()); Lname.append("nhtus = 1, subdt = 3")

exp0 = experiment(get_param(subdt = 3, nhtus = 2), Vstar = [1000, 1000])
Lout.append(exp0.launch()); Lname.append("nhtus = 2, subdt = 3")

for out, name, color in zip(Lout,Lname, ["k", "b", "r", "g"]):
  plt.plot(out[:100], label = name)
  plt.ylabel("Qout")
  plt.xlabel("basic time unit")
  plt.legend()
```

<h3> Conclusion </h3>
The relationship between volume / dt and corresponding discharge is different. The following experiment may help to understand it to be able to perform more astuciously the experiments.

```
# Experiment 2
"""
Force the "star" HTUs with a total Qin = 1000 equally shared and see the content of water of the HTUs at equilibrium.
"""
Lout = []; Lname = []

exp0 = experiment(get_param(), Qin = 1000)
Lout.append(exp0.launch()); Lname.append("nhtus = 1, subdt = 1")
print("nhtus = 1, subdt = 1")
print("Vstar:", exp0.Vstar, "V:", exp0.V)

exp0 = experiment(get_param(nhtus = 2), Qin = 500)
Lout.append(exp0.launch()); Lname.append("nhtus = 2, subdt = 1")
print("nhtus = 2, subdt = 1")
print("Vstar:", exp0.Vstar, "V:", exp0.V)

exp0 = experiment(get_param(subdt = 3), Qin = 1000)
Lout.append(exp0.launch()); Lname.append("nhtus = 1, subdt = 3")
print("nhtus = 1, subdt = 3")
print("Vstar:", exp0.Vstar, "V:", exp0.V)

exp0 = experiment(get_param(subdt = 3, nhtus = 2), Qin = 500)
Lout.append(exp0.launch()); Lname.append("nhtus = 2, subdt = 3")
print("nhtus = 2, subdt = 3")
print("Vstar:", exp0.Vstar, "V:", exp0.V)

color = ["k", "b", "r", "g"]
for i in [0,1]:
  plt.plot(Lout[i][:200], label = Lname[i])
plt.ylabel("Qout")
plt.xlabel("basic time unit")
plt.legend()
plt.show()
for i in [2,3]:
  plt.plot(Lout[i][:200], label = Lname[i])
plt.ylabel("Qout")
plt.xlabel("basic time unit")
plt.legend()
plt.title("")
plt.show()
```

<h3> Conclusion </h3>

In these cases, neither the time step nor the time step change the experiment.  
We have tcst = 0.1 and topoindex = 300 and the discharge at equilibrium is = Qin = 1000. For one "star" HTUs and for the linear HTUs, we have : 

$Veq = Qeq * tcst * topoindex / dt$ (and $dt = 1 / subdt$)

So in this case the volume at equilibrium is 30000 for subdt = 1, 90000 for subdt = 3. If nhtus >1, this volume is equally shared between the shared HTUs.

```
# Experiment 3
"""
Same experiment but with the equilibrium volume at Qin = 1000 and setting Qin = 0.
"""
Lout = []; Lname = []

exp0 = experiment(get_param(), V = [30000, 30000], Vstar = [30000])
Lout.append(exp0.launch()); Lname.append("nhtus = 1, subdt = 1")

exp0 = experiment(get_param(nhtus = 2), V = [30000, 30000], Vstar = [15000,15000])
Lout.append(exp0.launch()); Lname.append("nhtus = 2, subdt = 1")

exp0 = experiment(get_param(subdt = 3), V = [30000, 30000], Vstar = [30000])
Lout.append(exp0.launch()); Lname.append("nhtus = 1, subdt = 3")

exp0 = experiment(get_param(subdt = 3, nhtus = 2), V = [30000, 30000], Vstar = [15000,15000])
Lout.append(exp0.launch()); Lname.append("nhtus = 2, subdt = 3")

for i in [0,1]:
  plt.plot(Lout[i][:200], label = Lname[i])
plt.ylabel("Qout")
plt.xlabel("basic time unit")
plt.legend()
plt.show()

for i in [2,3]:
  plt.plot(Lout[i][:200], label = Lname[i])
plt.ylabel("Qout")
plt.xlabel("basic time unit")
plt.legend()
plt.show()
```


<h3>Conclusion</h3>
Once again we see that in this case neither the time step nor the number of HTUs seems to have an influence on the change of state.

The aim of this section is to evaluate if there can be limit for the dt description. It contains two type of experiments :  

* Experiment with the same dt and different nhtus
* Experiment with different dt and same nhtus

```
# Experiment 4
"""
We take subdt = 3. The volume are set at equelibrium for a Qin = 1000.
The simulation have Qin = 0 -> let see how the htus will empty
"""

Lout = []; Lname = []

exp0 = experiment(get_param(subdt = 3), V = [30000, 30000], Vstar = [30000])
Lout.append(exp0.launch()); Lname.append("nhtus = 1, subdt = 3")

exp0 = experiment(get_param(subdt = 3,nhtus = 2), V = [30000, 30000], Vstar = [15000,15000])
Lout.append(exp0.launch()); Lname.append("nhtus = 2, subdt = 3")

exp0 = experiment(get_param(subdt = 3, nhtus = 4), V = [30000, 30000], Vstar = [7500]*4)
Lout.append(exp0.launch()); Lname.append("nhtus = 4, subdt = 3")

exp0 = experiment(get_param(subdt = 3, nhtus = 10), V = [30000, 30000], Vstar = [3000]*10)
Lout.append(exp0.launch()); Lname.append("nhtus = 10, subdt = 3")

for i in [0,1,2,3]:#range(len(Lout))]:
  plt.plot(Lout[i][:200], label = Lname[i])
plt.ylabel("Qout")
plt.xlabel("basic time unit")
plt.legend()
plt.show()
```

<h3> Conclusion </h3>
All these configurations lead to an identical result.
